# MoReVis preprocessing datasets

This notebook will preprocess our datasets and save it for be displayed on the visualization interface.

```python
import sys
import os
module_path = os.path.abspath(os.path.join('../scripts'))
sys.path.append(module_path)

from morevis import MoReVis, compute_intersections_info
from datasets import load_dataset
```

## 2. Wildtrack


### 2.1 Subset 1

```python
df = load_dataset("wildtrack_use_case")
df_result = MoReVis(df)
df_intersection = compute_intersections_info(df_result)
    
df_result.to_csv("../app/static/data/wildtrack_small.csv", index = False)
df_intersection.to_csv("../app/static/data/wildtrack_small_intersections.csv", index = False)
```

### 2.2 Subset 2

```python
df = load_dataset("wildtrack_use_case_2")
df_result = MoReVis(df)
df_intersection = compute_intersections_info(df_result)
    
df_result.to_csv("../app/static/data/wildtrack_small_2.csv", index = False)
df_intersection.to_csv("../app/static/data/wildtrack_small_2_intersections.csv", index = False)
```

## 3. HURDAT

```python
df = load_dataset("hurdat_use_case")
df_result = MoReVis(df, height_method="area_max")
df_intersection = compute_intersections_info(df_result)

df_result.to_csv("../app/static/data/hurdat.csv", index = False)
df_intersection.to_csv("../app/static/data/hurdat_intersections.csv", index = False)
```
